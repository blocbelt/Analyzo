```python
def visualize_building_type_distribution(data, title):
    """
    This function visualizes the distribution of apartments in new and old buildings across different locations.

    Parameters:
    data (DataFrame): The dataset containing apartment information.
    title (str): The title for the overall plot.

    The function segregates the data into two categories based on whether the buildings are new or old, and then plots
    the top 10 locations for each category. It uses count plots to represent the number of apartments.
    """
    # Filter data for new and old buildings
    new_buildings = data[data['new_building'] == 1]
    old_buildings = data[data['new_building'] == 0]

    # Create subplots
    fig, ax = plt.subplots(1, 2, figsize=(15, 10))

    # Plot for new buildings
    sns.countplot(y='location', data=new_buildings, order=new_buildings.location.value_counts().index[:10], ax=ax[0])
    ax[0].set_title("Number of Apartments in New Buildings")

    # Plot for old buildings
    sns.countplot(y='location', data=old_buildings, order=old_buildings.location.value_counts().index[:10], ax=ax[1])
    ax[1].set_title("Number of Apartments in Old Buildings")

    # Set overall title and layout adjustments
    fig.suptitle(title, fontsize=16)
    plt.tight_layout(pad=3.0)
    plt.show()
```

```python
visualize_building_type_distribution(df,"New And Old Building Apartments in Baku")
```

```python
def bill_of_sale(data, title):
    """
    This function generates two side-by-side bar plots comparing the number of apartments
    with and without a bill of sale in different locations.

    Parameters:
    data (DataFrame): A pandas DataFrame containing the dataset.
    title (str): The title for the overall figure.

    The function splits the data based on the presence of a bill of sale and then
    generates count plots for the top 10 locations for each category.
    """

    # Splitting the dataset based on the presence of a bill of sale
    with_bill_of_sale = data[data['has_bill_of_sale'] == 1]
    without_bill_of_sale = data[data['has_bill_of_sale'] == 0]

    # Setting up the matplotlib figure with two subplots
    fig, ax = plt.subplots(1, 2, figsize=(15, 10))

    # Plotting the data for apartments with a bill of sale
    sns.countplot(y="location", data=with_bill_of_sale,
                  order=with_bill_of_sale['location'].value_counts().index[:10], ax=ax[0])
    ax[0].set_title("Number of Apartments with Bill of Sale")

    # Plotting the data for apartments without a bill of sale
    sns.countplot(y='location', data=without_bill_of_sale,
                  order=without_bill_of_sale['location'].value_counts().index[:10], ax=ax[1])
    ax[1].set_title("Number of Apartments without Bill of Sale")

    # Setting the main title and adjusting layout
    fig.suptitle(title, fontsize=16)
    plt.tight_layout(pad=3.0)
    plt.show()
```

```python
bill_of_sale(df,"Building in Baku with and without Bill of Sale")
```

```python
def plot_room_price_relation(data):
    """
    Plots a categorical chart illustrating the relationship between the number of rooms and prices in Baku.

    Parameters:
    data (DataFrame): A Pandas DataFrame containing the columns 'rooms' and 'price'.

    Returns:
    None: The function directly displays the plot.
    """
    # Creating the categorical plot
    room_price_plot = sns.catplot(x="rooms", y="price", data=data)

    # Setting the size of the figure
    room_price_plot.fig.set_size_inches(3, 4)

    # Adding a title to the plot
    room_price_plot.fig.suptitle("Number of Rooms vs. Price in Baku", fontsize=12)

    # Setting axis labels
    room_price_plot.set_axis_labels("Number of Rooms", "Price (AZN)")

    # Formatting the y-axis labels to include commas in numbers for readability
    room_price_plot.ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: '{:,.0f}'.format(x)))

    # Displaying the plot
    plt.show()
```

```python
# Calculate IQR for each column
Q1 = df.quantile(0.20)
Q3 = df.quantile(0.85)
IQR = Q3 - Q1

# Filter out the outliers
df_no_outliers = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler, PolynomialFeatures
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet, BayesianRidge
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor, GradientBoostingRegressor, AdaBoostRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.gaussian_process import GaussianProcessRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
```

# Preprocessing
categorical_features = ['your_categorical_columns']

categorical_transformer = Pipeline(steps=[

```python
# Numerical Transformer
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),  # Replace 'mean' with 'median' or 'most_frequent' if needed
    ('scaler', StandardScaler())])
```

```python
# Preprocessor
preprocessor = ColumnTransformer(transformers=[('num', numerical_transformer, X.columns)])
```

```python
# Model dictionary
models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(),
    'Lasso Regression': Lasso(),
    'Elastic Net Regression': ElasticNet(),
    'Polynomial Regression': Pipeline(steps=[('poly', PolynomialFeatures(degree=2)),
                                             ('linear', LinearRegression())]),
    'SVR': SVR(),
    'Decision Tree Regressor': DecisionTreeRegressor(),
    'Random Forest Regressor': RandomForestRegressor(),
    'Extra Trees Regressor': ExtraTreesRegressor(),
    'Bayesian Ridge Regression': BayesianRidge(),
    'XGBoost': XGBRegressor(),
    'LightGBM': LGBMRegressor(),
    'CatBoost': CatBoostRegressor(verbose=0),
    'KNN Regressor': KNeighborsRegressor(),
    'Gradient Boosting Regressor': GradientBoostingRegressor(),
    'Gaussian Process Regression': GaussianProcessRegressor(),
    'AdaBoost Regressor': AdaBoostRegressor()
}
```

# Train, Evaluate, and Store Results:
Loop through the models to train and evaluate:

```python
# Dictionary to store results
results = {}

# Train and evaluate each model
for name, model in models.items():
    pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                               ('model', model)])
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    results[name] = {'R2 Score': r2, 'RMSE': rmse}
```

```python
# Output the results
for model, scores in results.items():
    print(f"{model}: R2 Score = {scores['R2 Score']}, RMSE = {scores['RMSE']}")
```

```python
r2_scores = {model: metrics['R2 Score'] for model, metrics in results.items()}
rmse_scores = {model: metrics['RMSE'] for model, metrics in results.items()}

plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.barh(list(r2_scores.keys()), list(r2_scores.values()))
plt.xlabel('R2 Score')
plt.title('Model R2 Score Comparison')

plt.subplot(1, 2, 2)
plt.barh(list(rmse_scores.keys()), list(rmse_scores.values()))
plt.xlabel('RMSE')
plt.title('Model RMSE Comparison')

plt.tight_layout()
plt.show()
```

```python
# Find the model with the highest R2 score and the lowest RMSE
best_r2_model = max(results, key=lambda key: results[key]['R2 Score'])
best_rmse_model = min(results, key=lambda key: results[key]['RMSE'])

print(f"Best Model based on R2 Score: {best_r2_model} with R2 Score = {results[best_r2_model]['R2 Score']}")
print(f"Best Model based on RMSE: {best_rmse_model} with RMSE = {results[best_rmse_model]['RMSE']}")


# R2 score:
best_model_name = best_r2_model
best_model = models[best_model_name]
```
